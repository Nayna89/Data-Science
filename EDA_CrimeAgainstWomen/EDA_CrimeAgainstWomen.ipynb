## 4. Data normalization

<a id=section401></a>
### 4.1 Droping irrelevant columns 

```python
caw = caw.drop(['Unnamed: 22'], axis=1)
caw = caw.drop(['Sub_Group_Name'], axis=1)
```

<a id=section402></a>
### 4.2 Standardize column headers

```python
#renameing column names to proper and small names
caw = caw.rename(index=str, columns={"Group_Name" : "Crime",
                              "Cases_Acquitted_or_Discharged": "acquitted_or_discharged", 
                              "Cases_charge_sheets_were_not_laid_but_Final_Report_submitted": "FR_without_chargesheets", 
                              "Cases_Declared_False_on_Account_of_Mistake_of_Fact_or_of_Law": "declared_false", 
                              "Cases_Investigated_Chargesheets+FR_Submitted": "investigated_CS+FR_Submitted", 
                              "Cases_not_Investigated_or_in_which_investigation_was_refused": "not_investigated_or_investigation_refused", 
                              "Cases_Pending_Investigation_at_Year_End": "pending_investigation_at_year_end", 
                              "Cases_Pending_Investigation_from_previous_year": "pending_investigation_from_previous_year", 
                              "Cases_Pending_Trial_at_Year_End": "pending_trial_at_year_end", 
                              "Cases_withdrawn_by_the_Govt_during_investigation": "withdrawn_by_the_govt_during_investigation"})

caw.columns = map(str.lower, caw.columns)                               
caw.dtypes                                                              # Confirm that all variable names are now lower case
```

### 4.3 Grouping 

```python
caw['crime'].unique()
```

```python
caw.groupby(['crime'])['cases_reported'].sum().sort_values(ascending = False)
```

We can see that 'Sati Prevention Act', 'Importation of Girls' and 'Indecent Representation of Women (Prohibition) Act'
crimes have very less number of total number of crimes, so lets group them together and name it as 'Others'

Also we can see there is a crime name called 'Total Crime Against Women' which is assumed to be total of other crimes, so we will be removing these entries while anaysing further, as they may result in wrong analysis.  

```python
cawnew = caw.copy()
cawnew['crime'] = cawnew['crime'].replace({'Sati Prevention Act': 'Others', 'Importation of Girls': 'Others', 
                                   'Indecent Representation of Women (Prohibition) Act': 'Others'})
cawnew = cawnew[~cawnew['crime'].isin(['Total Crime Against Women'])]
cawnew['crime'].unique()
```

```python
cawnew.groupby(['crime'])['cases_reported'].sum().sort_values(ascending = False)     #check total after changing the crime name to others
```

### 6.1. "AREA" -  Is Your State Safe For Women?

<a id=section6011></a>
#### 6.1.1 India map showing States with total number of crimes reported against women.

```python
#Reding data from States.csv which has list of States and their Longitude and Latitude Data. 
#This data was taken from online data sources.
states = pd.read_csv("https://raw.githubusercontent.com/Nayna89/Data-Science/master/EDA_CrimeAgainstWomen/States.csv") 

areatotalcases = cawnew.groupby('area_name').sum()[['cases_reported']] 
areacasesLL = areatotalcases.merge(states, left_on='area_name', right_on='States')
```

```python
areatotalcases = areatotalcases.reset_index(drop = False)
areatotalcases.columns = ['States', 'Total Cases Reported']
areatotalcases.sort_values(by = 'Total Cases Reported', ascending = False).head(10)
```

Above table represents top 10 unsafe States - states which have more number of cases reported. 

```python
areatotalcases.sort_values(by = 'Total Cases Reported').head(10)
```

Above table represents top 10 safe States - states which have less number of cases reported. 

```python
#!pip install plotly
import plotly
from scipy.interpolate import spline
from numpy import array
import matplotlib as mpl

# for plots
from matplotlib import cm
from matplotlib.dates import date2num
from mpl_toolkits.basemap import Basemap

# for statistical graphs
from mpl_toolkits.basemap import Basemap

sns.set(style='white')

# Creating a function to plot the population data on real India map

def plot_map(sizes, colorbarValue):

    plt.figure(figsize=(9,10))
    f, ax = plt.subplots(figsize=(9, 10))

    # Setting up Basemap
    map = Basemap(width=5000000, height=3500000, resolution='l', projection='aea', llcrnrlon=69,
                  llcrnrlat=6, urcrnrlon=99, urcrnrlat=36, lon_0=78, lat_0=20, ax=ax)
                  
    # draw map boundaries
    map.drawmapboundary()
    map.drawcountries()
    map.drawcoastlines()

    # plotting states on map using previously derived coordinates
    x, y = map(array(areacasesLL["Longitude"]), array(areacasesLL["Latitude"]))
    cs = map.scatter(x, y, s=sizes/50, marker="o", c=sizes, cmap=cm.Dark2, alpha=0.5)

    # adding colorbar
    cbar = map.colorbar(cs, location='right',pad="5%")
    cbar.ax.set_yticklabels(colorbarValue)

    plt.show()
    
population_sizes = areacasesLL["cases_reported"]

colorbarValue = np.linspace(areacasesLL["cases_reported"].min(), areacasesLL["cases_reported"].max(), 
                            num=10)

colorbarValue = colorbarValue.astype(int)


plot_map(population_sizes, colorbarValue)
```

__Observation__ <br>
We can see that cases reported is widey spread all over India, 
under which AP, UP, Maharashtra, MP, West Bengal and Rajasthan are top states and are less safe for women.

### 6.4. "TRUST" - Are there any loopholes in legislation?

<a id=section6041></a>
#### 6.4.1. Different types of cases Withdrawn comparison

```python
# Make data:
cwsum =  cawnew['cases_compounded_or_withdrawn'].sum()
wgdisum =  cawnew['withdrawn_by_the_govt_during_investigation'].sum()
cgsim = cawnew['cases_withdrawn_by_the_govt'].sum()
totalw = cawnew['cases_reported'].sum()


group_names=['cases_withdrawn_by_the_govt', 'withdrawn_by_the_govt_during_investigation','cases_compounded_or_withdrawn']
group_size=[cgsim, wgdisum, cwsum]
 
fig, ax = plt.subplots()
mypie, _ = ax.pie(group_size, radius=1.3)
plt.setp(mypie, width=0.3)
#plt.title('Donut plot')
# show it
plt.show()


print('Green - cases_compounded_or_withdrawn = ', round(100*(cwsum/totalw), 1), '%')
print('Blue - cases_withdrawn_by_the_govt = ', round(100*(cgsim/totalw), 1), '%')
print('Orange - withdrawn_by_the_govt_during_investigation = ', round(100*(wgdisum/totalw), 1), '%')
```

__Observation__ - <br>
1. It says, Govt. don’t investigate much before withdrawing a case.<br/>
2. 23% of cases were withdrawn by Govt. rest were compounded.

### 6.5. RESULT - What is the outcome of cases reported?

<a id=section6051></a>
#### 6.5.1. Case life cyle

```python
from pandas.plotting import parallel_coordinates

#Considering only top reported crimes - CASES UNDER WOMEN ASSAULTED for cases reported more than 500 
f = (cawnew.loc[cawnew['crime'].isin(['Dowry Deaths', 'Cruelty by Husband and Relatives', 'Rape'])]
     .loc[cawnew['cases_reported'] >500]
     .loc[:, ['cases_reported', 'cases_sent_for_trial', 'cases_chargesheeted',
              'investigated_cs+fr_submitted', 'cases_convicted', 'acquitted_or_discharged']]
    )


f['Crime'] = cawnew['crime']
f = f.sample(100)                                    #Taking 100 sample
plt.figure(figsize=(15,7))
parallel_coordinates(f, 'Crime', color=('#32CD32','#FFFF00', '#DB7093'))
plt.show()
```

__Observation__ - <br> 
For cases reported under women assaulted, charge sheets were prepared, cases got investigated and final report got submitted, even sent for trial but at last most of the cases got acquitted or accused got discharged instead of convicted. 

Some reasons for this may be – 
1. Lack of proof submission in trials 
2. Improper investigation
3. Evidence gets erased
4. Poverty vs. wealth

<a id=section6052></a>
#### 6.5.2. Cases final state - discarded or convicted? 

```python
#Suming all field values which comes under cases discarded
cawnew["total_cases_discarded"]  = cawnew['declared_false'] + cawnew['withdrawn_by_the_govt_during_investigation']
+ cawnew['not_investigated_or_investigation_refused'] + cawnew['cases_withdrawn_by_the_govt']
+ cawnew['cases_compounded_or_withdrawn'] + cawnew['acquitted_or_discharged']

#Splitting the dataset into 2, so have clear visualization 
group1 = cawnew[cawnew['crime'].isin(['Cruelty by Husband and Relatives', 
                                    'Dowry Deaths', 
                                   'Rape',
                                   'Sexual harassment'])]

group2 = cawnew[cawnew['crime'].isin(['Molestation',
                                   'Sati Prevention Act',
                                    'Importation of Girls', 
                                    'Kidnapping & Abduction - Women & Girls'
                                   'Indecent Representation of Women (Prohibition) Act',
                                   'Immoral Traffic (Prevention) Act'])]
plt.figure(figsize=(22,6))
sns.lmplot(x="cases_convicted", y="total_cases_discarded", hue='crime', palette = 'husl', data=group1)
plt.ylim(-1000, max(cawnew['total_cases_discarded']) + 1000)
plt.xlim(-1000, max(cawnew['cases_convicted']) + 1000)
plt.xlabel("Cases Convicted")
plt.ylabel("Total Cases Discarded")
plt.show()

plt.figure(figsize=(22,6))
sns.lmplot(x="cases_convicted", y="total_cases_discarded", hue='crime', palette = 'husl', data=group2)
plt.xlabel("Cases Convicted")
plt.ylabel("Total Cases Discarded")
plt.ylim(-1000, max(cawnew['total_cases_discarded']) + 1000)
plt.xlim(-1000, max(cawnew['cases_convicted']) + 1000)
plt.show()
```

__Observation__ - <br>

1. We can notice, “Cruelty by Husband and Relatives” is widely spread and we can notice that mostly these cases  are discarded rather than convicted.<br>
2. Also cases under “Immoral Traffic Act” is widely spread towards cases discarded.

## 7. Conclusion & Actionable Insights

Under 5 categories of questions, we have below conclusion and their respective actianable insights.

__1. Area__ - 

__Conclusion__ - AP, UP and MP are most unsafe areas for women. Whereas Union Territories are the safest areas.<br>

<br>
__AI__ - The incorporation of stringent laws and stricter punishments are important to deter people from committing such crimes.

__3. Time__ - 

__Conclusion__ - Year on year, total crime against women is increasing. And also pending cases are not reducing.
<br>
__AI__ - An improvement in court management should certainly be considered.

__4. Trust__ - 

__Conclusion__ - Most of the cases are not being sent for trial itself, and this leads to increase in backlogs.
<br>
__AI__ - More and more fast track court should establish

__5. Result__ - 

__Conclusion__ - Mostly cases are discarded rather than convicted.
<br>
__AI__ - More concentration and awareness should be spread to collect proper proofs for submission in trials.
